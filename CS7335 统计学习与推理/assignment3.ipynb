```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

feature_dim = 2048

A2R_source_df = pd.read_csv('OfficeHome/features/Art_Art.csv')
A2R_target_df = pd.read_csv('OfficeHome/features/Art_RealWorld.csv')

A2R_source_features = A2R_source_df.iloc[:, :feature_dim].values
A2R_source_labels = A2R_source_df.iloc[:, feature_dim:].values.reshape(-1).astype(int)
print(f"A2R Source Domain Sample number: {A2R_source_labels.shape}")
A2R_target_features = A2R_target_df.iloc[:, :feature_dim].values
A2R_target_labels = A2R_target_df.iloc[:, feature_dim:].values.reshape(-1).astype(int)
print(f"A2R Target Domain Sample number: {A2R_target_labels.shape}")
```

```python
C2R_source_df = pd.read_csv('OfficeHome/features/Clipart_Clipart.csv')
C2R_target_df = pd.read_csv('OfficeHome/features/Clipart_RealWorld.csv')

C2R_source_features = C2R_source_df.iloc[:, :feature_dim].values
C2R_source_labels = C2R_source_df.iloc[:, feature_dim:].values.reshape(-1).astype(int)
print(f"C2R Source Domain Sample number: {C2R_source_labels.shape}")
C2R_target_features = C2R_target_df.iloc[:, :feature_dim].values
C2R_target_labels = C2R_target_df.iloc[:, feature_dim:].values.reshape(-1).astype(int)
print(f"C2R Target Domain Sample number: {C2R_target_labels.shape}")
```

```python
A2R_source_label_set = list(set(A2R_source_labels))
A2R_source_sample_num = [0 for _ in range(len(A2R_source_label_set))]
for label in A2R_source_labels:
    A2R_source_sample_num[label] += 1
canvas = plt.figure()
plt.cla()
plt.bar(A2R_source_label_set, A2R_source_sample_num)
plt.xlabel("Labels")
plt.ylabel("Frequency")
plt.title("Art->RealWorld Source Domain Sample Distribution")
plt.xlim(-1,65)
```

```python
A2R_target_label_set = list(set(A2R_target_labels))
A2R_target_sample_num = [0 for _ in range(len(A2R_target_label_set))]
for label in A2R_target_labels:
    A2R_target_sample_num[label] += 1
canvas = plt.figure()
plt.cla()
plt.bar(A2R_target_label_set, A2R_target_sample_num)
plt.xlabel("Labels")
plt.ylabel("Frequency")
plt.title("Art->RealWorld Target Domain Sample Distribution")
plt.xlim(-1,65)
```
